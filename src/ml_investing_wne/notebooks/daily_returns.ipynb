```python
trades = trades[['datetime', 'close', 'barrier_touched','barrier_touched_date', 'top_barrier','bottom_barrier', 'transaction', 'budget']]
```

```python
config.cost = 0.0025
trades['cost_exit'] = 0
trades['cost_entry'] = 0
```

```python
for i in trades.index:
    trades.loc[i, 'price_entry'] = trades.loc[i, 'close']

    # prices
    if trades.loc[i, 'transaction'] in ['sell', 'buy']:
        if trades.loc[i, 'barrier_touched'] == 'top':
            trades.loc[i, 'price_exit'] = trades.loc[i, 'top_barrier']
        elif trades.loc[i, 'barrier_touched'] == 'bottom':
            trades.loc[i, 'price_exit'] = trades.loc[i, 'bottom_barrier']
        elif trades.loc[i, 'barrier_touched'] == 'vertical':
            trades.loc[i, 'price_exit'] = trades.loc[i+1, 'close']
    elif trades.loc[i, 'transaction'] == 'No trade':
        trades.loc[i, 'price_exit'] = trades.loc[i+1, 'close']

    # cost
    if i == 0 and trades.loc[i, 'transaction'] in ['sell', 'buy']:
        trades.loc[i, 'cost_entry'] = config.cost
    elif i > 0:
        # if the transaction is different than the previous one
        if trades.loc[i, 'transaction'] != trades.loc[i-1, 'transaction']:
            # if previously there was no transaction, there is no exit cost
            if trades.loc[i-1, 'transaction'] == 'No trade':
                trades.loc[i, 'cost_exit'] = 0
            else:
                trades.loc[i-1, 'cost_exit'] = config.cost
            if trades.loc[i, 'transaction'] in ['sell', 'buy']:
                trades.loc[i, 'cost_entry'] = config.cost
        if i == trades.index[-1] and trades.loc[i, 'transaction'] in ['sell', 'buy']:
            trades.loc[i, 'cost_exit'] = config.cost

    

```
