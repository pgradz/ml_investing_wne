```python
import importlib
build_model = getattr(importlib.import_module(f'ml_investing_wne.tf_models.lstm'),'build_model')
model = build_model(input_shape=(10, 41), nb_classes=config.nb_classes)
```

```python
model.fit(train_dataset_2, epochs=1, verbose=2)
```
